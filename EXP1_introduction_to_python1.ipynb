### if-else statements
`if condition:`
   >  `statements if condition is TRUE`<br>

   `elif condition:` <br>
   > `statements` <br>

`else:`<br>
   >  `statements`

```python
marks = 9
if marks > 90:
  print('Congrats! A!')
elif marks > 80:
  print("Alright!")
else:
  print('Need to do better!')
```

### Loops
### for loop
`for i in range(N):`

```python
for a in range(2,10,2):
  print(a)  # indexing starts from 0 in python
```

### while loop
`while condition:` <br>
> `statements`

```python
a = 5

while a <= 10:
  print(a)
  a = a + 1
```

### List (like Array in C programming)
`list_name = [val_1, val_2, val_3, .......]`

```python
marks = [88, 56, 94, 92, 76]

print(len(marks))


for n in marks:
  print("The marks of Student is {}".format(n)) #{} placeholder in python
```

### Common List operations

```python
marks = [88, 56, 92, 82, 94]
print("Before changing any element                --> ", marks)

marks.append(99)
print("After appending one element                --> ", marks)

marks.remove(94)
print("After removing one element                 --> ", marks)

marks.insert(-1, 71)
print("After inserting '71' at index '1'          --> ", marks)

marks.pop(2)
print("After popping out the element at index '2' --> ", marks)
```
